```python
import easyocr
from PIL import Image
import io
import openpyxl

def extract_text_from_image(image_path):
    reader = easyocr.Reader(['en'])
    image = Image.open(image_path)
    img_bytes = io.BytesIO()
    image.save(img_bytes, format='PNG')
    img_bytes.seek(0)
    result = reader.readtext(img_bytes.getvalue(), detail=0)
    text = "\n".join(result)
    print(text)
    return text

def extract_table_from_text(text):
    # Implement logic to parse the OCR text into a table format
    # For demonstration, we assume the text is already in a table format
    table = [row.split() for row in text.splitlines()]
    print(table)
    return table

def save_table_to_excel(table, output_path):
    workbook = openpyxl.Workbook()
    sheet = workbook.active
    for row_idx, row in enumerate(table, 1):
        for col_idx, cell in enumerate(row, 1):
            sheet.cell(row=row_idx, column=col_idx, value=cell)
    workbook.save(output_path)

# Example usage
image_path = "Data/input.jpeg"
output_path = "/content/drive/MyDrive/Internship/Output.xlsx"

text = extract_text_from_image(image_path)
table = extract_table_from_text(text)
save_table_to_excel(table, output_path)

print("Extracted table from image and saved to", output_path)
```

```python
from img2table.ocr import TesseractOCR
from img2table.document import Image
src = "Data/input.jpeg"
# Check if src is None
if src is None:
    print("Error: src is None")
    exit()

# Check if src is a valid file
import os
if not os.path.exists(src) or not os.path.isfile(src):
    print("Error: src is not a valid file")
    exit()

# Instantiation of OCR
oc= TesseractOCR(n_threads=4, lang="eng")

# Instantiation of document, either an image or a PDF
doc = Image(src)

# Table extraction
extracted_tables = doc.extract_tables(ocr=oc,

                                      implicit_rows=False,
                                      borderless_tables=False,
                                      min_confidence=50)
```

### Gemini Model Performance (Working)

```python
import google.generativeai as genai
import os
genai.configure(api_key=os.getenv("GOOGLE_API_KEY"))



def input_image_setup(file_location, type='image/jpeg'):
    # Check if the file location is valid
    if file_location:

            # Open the file in binary mode and read it into bytes
        with open(file_location, 'rb') as f:
                bytes_data = f.read()

        image_parts = [
                {
                    "mime_type": type,  # Get the mime type of the uploaded file
                    "data": bytes_data
                }
            ]

        input_prompt="""
 You are an expert in reading handwriting in english. There is a table present in the provided image. Read the image and extract the table data which include Names , phone number and email . Try your best to get the best possible result. The handwriting will in different in every rows. Identify the bounding lines to extract information row wise.
               The table should be in the following format:
               Format should not be changed anyhow.
              [[Column1,Column2,...],[Data1,Data2,Data3...],[]...]
            """

        generation_config = {
  "temperature": 0.1,
  "top_p": 1,
  "top_k": 1,
  "max_output_tokens": 2048,
    }
        model=genai.GenerativeModel('gemini-pro-vision',generation_config=generation_config)
        response=model.generate_content([input_prompt,image_parts[0]])

        print(response.text)
        return response.text

Raw=input_image_setup('Data/input.jpeg')
```

```python
import openpyxl
from openpyxl.utils.cell import get_column_letter
import pandas as pd

def data_to_xlsx(data, columns, filename):
    # Create a pandas DataFrame from the data
    df = pd.DataFrame(data, columns=columns)
    print(df)
    # Create a new workbook and select the active worksheet
    workbook = openpyxl.Workbook()
    sheet = workbook.active

    # Write headers to the first row
    for col_num, header in enumerate(columns, 1):
        sheet.cell(row=1, column=col_num, value=header)

    # Write data to subsequent rows
    for row_num, row_data in enumerate(df.values, 2):
        for col_num, value in enumerate(row_data, 1):
            sheet.cell(row=row_num, column=col_num, value=value)

    # Save the workbook to a file
    workbook.save(filename)



data_to_xlsx(d, c[0], "output/output.xlsx")
```
